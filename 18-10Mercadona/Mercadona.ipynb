# Agrupacin

Cuntos productos hay por categora? Ordena de menos a ms.

```python
productos_por_cat = df["Category"].value_counts().sort_values(ascending=True)
print("Productos por categora (de menos a ms):")
print(productos_por_cat)
print("Guillermo Garca Hernndez  2 DAW")
```

Cuntas categoras distintas hay? Usa el mtodo nunique()


```python
num_categorias = df["Category"].nunique()
print("Categoras distintas:", num_categorias)
print("Guillermo Garca Hernndez  2 DAW")
```

Cul es el nmero de descuentos que hay por categora?

```python
descuentos_por_cat = (
    df.assign(tiene_descuento=df["discount_price"].notna())
      .groupby("Category")["tiene_descuento"]
      .sum()               
      .sort_values(ascending=False)
)
print("Nmero de productos con descuento por categora:")
print(descuentos_por_cat)
print("Guillermo Garca Hernndez  2 DAW")
```

Cules son las 5 categoras con mayor precio medio?

```python
top5_media_precio = (
    df.groupby("Category")["price"]
      .mean()
      .sort_values(ascending=False)
      .head(5)
)
print("Top 5 categoras por precio medio:")
print(top5_media_precicio := top5_media_precio)  
print("Guillermo Garca Hernndez  2 DAW")
```

Devuelve el producto con menor precio por cada categora

```python
idx_min_por_cat = df.groupby("Category")["price"].idxmin()
productos_min_por_cat = df.loc[idx_min_por_cat, ["Category", "name", "subtitle", "price", "discount_price"]]
print("Producto ms barato por categora:")
print(productos_min_por_cat.sort_values("Category"))
print("Guillermo Garca Hernndez  2 DAW")
```

Agrupa por categora y por el descuento (slo los que tengan), y casa la media de precio.

```python
con_descuento = df[df["discount_price"].notna()]
media_precio_con_descuento = (
    con_descuento
      .groupby("Category")["price"]
      .mean()
      .sort_values(ascending=False)
)
print("Media de precio (solo productos con descuento) por categora:")
print(media_precio_con_descuento)
print("Guillermo Garca Hernndez  2 DAW")
```

Agrupa por subtitulo y cuenta cuntos productos tienen el mismo subttulo.

```python

conteo_subtitle = df["subtitle"].value_counts(dropna=False)
print("Productos por subtitle (incluye NaN):")
print(conteo_subtitle.head()) 
print("Total de subttulos distintos (incluyendo NaN):", conteo_subtitle.shape[0])
print("Guillermo Garca Hernndez  2 DAW")
```

Cuntos productos, por categora tienen descuentos y cuantos no? Crea una nueva columna que se llame "tiene_descuento", de tipo booleano.

Tras ello, realiza el agrupamiento.

```python
df_td = df.assign(tiene_descuento=df["discount_price"].notna())
descuentos_por_cat = (
    df_td.groupby(["Category", "tiene_descuento"])
         .size()
         .unstack(fill_value=0)  
         .rename(columns={False: "sin_descuento", True: "con_descuento"})
         .sort_values(by="con_descuento", ascending=False)
)
print("Conteo por categora (con/sin descuento):")
print(descuentos_por_cat)
print("Guillermo Garca Hernndez  2 DAW")
```

Saca una diagrama de lneas del precio. Qu informacin se presenta en cada eje?

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(9,4))
df["price"].plot(kind="line")
plt.title("Precio por ndice de fila (producto)")
plt.xlabel("ndice (fila del DataFrame)")
plt.ylabel("Precio ()")
plt.grid(True)
plt.tight_layout()
plt.show()
print("Guillermo Garca Hernndez  2 DAW")
```

Saca un histograma del precio, y del descuento, en el mismo grfico.

```python
plt.figure(figsize=(8,5))
plt.hist(df["price"].dropna(), bins=30, alpha=0.6, label="price")
plt.hist(df["discount_price"].dropna(), bins=30, alpha=0.6, label="discount_price")
plt.title("Histogramas: precio vs precio con descuento")
plt.xlabel("Precio ()")
plt.ylabel("Frecuencia")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
print("Guillermo Garca Hernndez  2 DAW")
```

Saca un diagrama de barras con el nmero de productos que tienen descuento por categora.

Para hacer este ejercicio, primero filtra aquellos productos que tengan descuento. Este resultado lo puedes almacenar en otra variable para mejorar la legibilidad.

Despus agrupa por categora y saca el diagrama.

```python
import matplotlib.pyplot as plt
# Filtrar los productos con descuento (no vacos)
df_discount = df[df["discount_price"].notna()]

# Contar productos por categora
count_by_category = df_discount["Category"].value_counts()

# Diagrama de barras
count_by_category.plot(kind="bar", color="orange", figsize=(6,3))
plt.title("Productos con descuento por categora")
plt.xlabel("Categora")
plt.ylabel("Nmero de productos")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.show()
print("Guillermo Garca Hernndez  2 DAW")
```
